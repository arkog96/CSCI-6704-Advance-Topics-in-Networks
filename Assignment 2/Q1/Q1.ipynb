```python
##CSCI 6704 Assignment 2 Question 1
##Arka Ghosh 
##BannerID: B00911033

import matplotlib.pyplot as plt
import numpy as np


def Unipolar(string):
    xs = []
    ys = []
    for i, b in enumerate("".join(str(e) for e in string)):
        xs.extend([i, i+1])
        #print(xs)
        ys.extend([0,0] if b== '0' else [1,1])
        #print(ys)
    plt.figure(figsize=(10,4))
    plt.grid()
    plt.plot(xs,ys)
    plt.title("Unipolar", y=-0.2, fontsize = 14)
    for i,b in enumerate(string):
        plt.text(i+.5,1.2,b)
    plt.savefig("Unipolar.jpg")
    plt.show()
    

def NRZ(string):
    xs = []
    ys = []
    for i, b in enumerate("".join(str(e) for e in string)):
        xs.extend([i, i+1])
        #print(xs)
        ys.extend([-1,-1] if b== '0' else [1,1])
        #print(ys)
    plt.figure(figsize=(10,4))
    plt.grid()
    plt.plot(xs,ys)
    plt.title("NRZ", y=-0.2, fontsize = 14)
    for i,b in enumerate(string):
        plt.text(i+.5,1.2,b)
    plt.savefig("NRZ.jpg")
    plt.show()
    

def Manchester(string):
    xs = []
    ys = []
    for i, b in enumerate("".join(str(e) for e in string)):
        xs.extend([i, i+0.5, i+0.5, i+1])
        ys.extend([1,1,-1,-1] if b== '0' else [-1,-1,1,1])
    plt.figure(figsize=(10,4))
    plt.grid()
    plt.plot(xs,ys)
    plt.title("Manchester", y=-0.2, fontsize = 14)
    for i,b in enumerate(string):
        plt.text(i+.5,1.2,b)
    plt.savefig("Manchester.jpg")
    plt.show()

bin_string = input("Input :")
bin_input = list(map(int, bin_string))
Unipolar(bin_input)
NRZ(bin_input)
Manchester(bin_input)
```
